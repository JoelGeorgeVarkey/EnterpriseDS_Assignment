```python
# importing Libraries
import pandas as pd
import missingno as msno
import plotly.express as px
import numpy as np

%matplotlib inline
```

```python
# Link for Data
Data_URL_Link = r"https://covid.ourworldindata.org/data/owid-covid-data.csv"
```

```python
# Loading data

df_country_info = pd.read_csv(Data_URL_Link, sep = ',')
# df_country_info.head(5)
```

```python
# Collecting data of 3 countries
df_GER = df_country_info[df_country_info['location']=='Germany']
df_GER =  df_GER.reset_index(drop=True)
df_US = df_country_info[df_country_info['location']=='United States']
df_US =  df_US.reset_index(drop=True)
df_IND = df_country_info[df_country_info['location']=='India']
df_IND =  df_IND.reset_index(drop=True)
```

```python
# Taking boolean ouptut to combine
b_GER = df_country_info['location']=='Germany'
b_US = df_country_info['location']=='United States'
b_IND = df_country_info['location']=='India'
```

```python
# Combining 3 countries
b_all = b_US | b_GER | b_IND

b_all
```

```python
# All 3 countries added to new variable
df_my_list = df_country_info[b_all]
```

```python
# Removing NaN
df_my_list = df_my_list.fillna(0)
```

```python
# Finding Number of cases per population
df_my_list['total_cases_per_population']=df_my_list['total_cases']/df_my_list['population']
```

```python
# Plotting the result as output
fig = px.line(df_my_list, x="date", y="total_cases_per_population",color = 'location', title='Covid cases')
fig.show()
```
