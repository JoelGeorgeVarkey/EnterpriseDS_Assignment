```python
# importing Libraries
import pandas as pd
import missingno as msno
import plotly.express as px
import numpy as np

import plotly.io as pio
pio.renderers

%matplotlib inline
```

```python
# Link for Data
Data_URL_Link = r"https://covid.ourworldindata.org/data/owid-covid-data.csv"
```

```python
# Loading data

df_country_info = pd.read_csv(Data_URL_Link, sep = ',')
# df_country_info.head(5)
```

```python
# Collecting data of 3 countries
df_GER = df_country_info[df_country_info['location']=='Germany']
df_GER =  df_GER.reset_index(drop=True)
df_US = df_country_info[df_country_info['location']=='United States']
df_US =  df_US.reset_index(drop=True)
df_IND = df_country_info[df_country_info['location']=='India']
df_IND =  df_IND.reset_index(drop=True)

df_GER
```

```python
# Adding data of 3 countries to new variable

# df_GER.loc[len(df_GER.index)] = [df_GER] 
# df_all = df_GER.append(df_GER, ignore_index = True)
df_all = pd.concat([df_GER, df_US, df_IND], ignore_index = True)
df_all.reset_index()

df_all
```

```python
# Taking the columns that are only necessary
df_all_reduce = df_all[["date","location","people_vaccinated", "population"]]
df_all_reduce
```

```python
# Removing NaN
# df_all_reduce = df_all_reduce.fillna(0)
# Not running this line because the graph of India looks not so good because of some breaks as it goes to zero.
```

```python
# TO find vaccination rate over time
df_all_reduce['vaccination_rate_over_time']=df_all_reduce['people_vaccinated']/df_all_reduce['population']
```

```python
# Plotting the result as output
fig = px.line(df_all_reduce, x="date", y="vaccination_rate_over_time",color = 'location', title='Covid cases')
fig.show()
```
